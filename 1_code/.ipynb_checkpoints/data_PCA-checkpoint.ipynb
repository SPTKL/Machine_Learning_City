# Feature selection

```python
#import modules we need to use for this session
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.formula.api as smf
from sklearn.decomposition import PCA
from sklearn.cross_validation import train_test_split
```

```python
X = df.drop(columns=['mean_income', 'user_id'])
Y = df.mean_income
```

```python
X.head()
```

```python
n=283 # how many eigenvectors we choose
from sklearn.decomposition import PCA
pca = PCA(n)
Xproj = pca.fit_transform(X)
eigenvalues = pca.explained_variance_
print(pca.explained_variance_ratio_)
plt.bar(np.arange(10), eigenvalues[0:10]);
plt.xlabel("Dimensionality")
plt.ylabel("Variance")
plt.show()
```

```python
plt.figure(2, figsize=(8, 6))
plt.clf()
# Plot the training points
plt.scatter(Xproj[:, 0], Xproj[:, 1], c=Y, cmap=plt.cm.cool)
plt.xlabel('First eigenvector')
plt.ylabel('Second eigenvector')
plt.show()
```
